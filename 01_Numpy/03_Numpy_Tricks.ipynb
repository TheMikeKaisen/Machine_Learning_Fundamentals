### np.sort

Return a sorted copy of an array.

https://numpy.org/doc/stable/reference/generated/numpy.sort.html

```python
# code
import numpy as np
a = np.random.randint(1,100,15) # random 15 numbers between 1 and 100 
a
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
# sorting in ascending order
# returns a numpy array
np.sort(a)
```

```python
# sorting in descending order
# np.sort doesn't have reverse=true parameter
np.sort(a)[::-1]
```

```python
# Sorting 2d 
# by rows, axis = 1
# by column, axis = 0
np.sort(b,axis=1)
```

```python
np.sort(b, axis = 0)
```

### np.where

The numpy.where() function returns the indices of elements in an input array where the given condition is satisfied.

https://numpy.org/doc/stable/reference/generated/numpy.where.html

```python
a
```

```python
# find all indices with value greater than 50
np.where(a>50)
```

```python
# replace all values > 50 with 0
# SYNTAX: np.where( condition, if true, if false )

# task: make all the elements which is greater that 60 to 0
np.where(a>60,0,a)
```

```python
# task: turn all the elements which are even to 0.
np.where(a%2 == 0,0,a)
```

### np.histogram
